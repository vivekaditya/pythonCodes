# Matplotlib tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[50,55,40,48,43]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather Chart')
lines = plt.plot(x,y,color="Green", linewidth=1.0, linestyle=":")
```
